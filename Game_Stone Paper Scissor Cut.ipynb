```python
def game_sps():
    point1 = 0
    point2 = 0

    while (point1 < 5) and (point2 < 5):  
        choice1 = input ('Choice of Player A: ')
        choice1 = choice1.lower().replace(' ','')
        print(choice1)
        choice2 = input ('Choice of Player B: ')
        choice2 = choice2.lower().replace(' ','')
        print(choice2)  
        if choice1 == 'stone':
            if choice2 =='stone':
                print('Draw')
            elif choice2 =='scissor':
                point1  = point1 + 1
                print('Point to Player A, Stone crushes the scissor', point1, 'Player B score is: ', point2)
            elif choice2 == 'paper':
                point2 = point2 + 1
                print('Point to Player B, Paper covers the stone','Player A score is: ', point1, 'Player B score is: ', point2 )
            else:
                print('Provide a valid choice for Player B')
        elif choice1 == 'paper':
            if choice2 =='paper':
                print('Draw')
            elif choice2 =='stone':
                point1  = point1 + 1
                print('Point to Player A, Paper covers the stone', point1, 'Player B score is: ', point2)
            elif choice2 == 'scissor':
                point2 = point2 + 1
                print('Point to Player B, Scissor cuts the paper','Player A score is: ', point1, 'Player B score is: ', point2 )
            else:
                print('Provide a valid choice for Player B')         
        elif choice1 == 'scissor':
            if choice2 =='scissor':
                print('Draw')
            elif choice2 =='stone':
                point2  = point2 + 1
                print('Point to Player B, Paper covers the stone', point1, 'Player B score is: ', point2)
            elif choice2 == 'paper':
                point1 = point1 + 1
                print('Point to Player A, Scissor cuts the paper','Player A score is: ', point1, 'Player B score is: ', point2 )
            else:
                print('Provide a valid choice for Player B')
        else:
            print('Provide a valid choice for Player A')
    if point1 < point2:
        print('Player B wins')
    else:
        print('Player A wins')
        
game_sps()
```
